```python
%matplotlib inline
import pandas as pd
import matplotlib.pyplot as plt
import datetime as dt
```

```python
df = pd.read_csv("D:/data2017/kc_house_train_data.csv")
```

```python
#Data manipulation
len(df)
```

```python
del df['id']
df['date'] = pd.to_datetime(df['date'])
```

```python
df = df.set_index('date')
```

```python
df2 = df.resample('M')['price'].agg(['mean', 'count']).reset_index()
```

```python
plt.plot(df2['date'], df2['mean'])
```

```python
df3 = df.resample('D')['price'].agg(['mean', 'count']).reset_index()
```

```python
plt.plot(df3['date'], df3['mean'])
```
